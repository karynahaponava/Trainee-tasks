**4.** Даны 2 массива A (8x3) и B (2x2). Найти строки в A, которые содержат элементы из каждой строки в B, независимо от порядка элементов в B

```python
a = np.random.randint(0, 10, (8, 3))
b = np.random.randint(0, 10, (2, 2))

print("Массив A:")
print(a)
print("Массив B:")
print(b)

check1 = np.isin(a, b[0]).any(axis=1)
check2 = np.isin(a, b[1]).any(axis=1)

mask = check1 & check2
result = a[mask]

print("Найденные строки в A:")
print(result)
```

**5.** Дана 10x3 матрица, найти строки из неравных значений (например строка [2,2,3] остается, строка [3,3,3] удаляется)

```python
mat = np.random.randint(0, 3, (10, 3))
mat[1] = [1, 1, 1]
mat[5] = [2, 2, 2]

print("Исходная матрица:")
print(mat)

all_equal_mask = (mat == mat[:, 0:1]).all(axis=1)

not_equal_mask = ~all_equal_mask
result = mat[not_equal_mask]

print("Строки, где не все элементы равны:")
print(result)
```

* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ 3.

```python
def diagonal_product_no_numpy(matrix):
    diagonal = [matrix[i][i] for i in range(min(len(matrix), len(matrix[0])))]
    
    product = 1
    for element in diagonal:
        if element != 0:
            product *= element
    
    return product

def diagonal_product_numpy(matrix):
    return np.prod(np.diag(matrix)[np.diag(matrix) != 0])

X_list = [[1, 0, 1],
          [2, 0, 2],
          [3, 0, 3], 
          [4, 4, 4]]

X_numpy = np.array([[1, 0, 1],
                    [2, 0, 2],
                    [3, 0, 3],
                    [4, 4, 4]])

print("=== Задача 1: Произведение ненулевых элементов на диагонали ===")
print(f"Матрица: {X_list}")
print(f"Без NumPy: {diagonal_product_no_numpy(X_list)}")  
print(f"С NumPy: {diagonal_product_numpy(X_numpy)}")      
```

* __Задача 2__: Даны два вектора x и y. Проверить, задают ли они одно и то же мультимножество.  
  Например, для x = np.array([1, 2, 2, 4]), y = np.array([4, 2, 1, 2]) ответ True.

```python
def task2_list(x, y):
    return sorted(x) == sorted(y)

def task2_numpy(x, y):
    return np.array_equal(np.sort(x), np.sort(y))

x_list = [1, 2, 2, 4]
y_list = [4, 2, 1, 2]
x_np = np.array(x_list)
y_np = np.array(y_list)

print(f"x = {x_list}, y = {y_list}")
print("List:", task2_list(x_list, y_list))
print("Numpy:", task2_numpy(x_np, y_np))
```

* __Задача 3__: Найти максимальный элемент в векторе x среди элементов, перед которыми стоит ноль. 
 Например, для x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) ответ 5.

```python
def task3_list(x):
    vals = []
    for i in range(len(x) - 1):
        if x[i] == 0:
            vals.append(x[i+1])

    return max(vals) if vals else None

def task3_numpy(x):
    zero_mask = (x == 0)

    vals = x[1:][zero_mask[:-1]]

    return vals.max() if vals.size > 0 else None

x_list = [6, 2, 0, 3, 0, 0, 5, 7, 0]
x_np = np.array(x_list)

print(f"x = {x_list}")
print("List:", task3_list(x_list))
print("Numpy:", task3_numpy(x_np))
```

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
def task4_list(x):
    if not x:
        return ([], [])

    vals = []
    counts = []

    curr_val = x[0]
    curr_count = 1

    for i in range(1, len(x)):
        if x[i] == curr_val:
            curr_count += 1
        else:
            vals.append(curr_val)
            counts.append(curr_count)
            curr_val = x[i]
            curr_count = 1

    vals.append(curr_val)
    counts.append(curr_count)

    return (vals, counts)

def task4_numpy(x):
    if x.size == 0:
        return (np.array([]), np.array([]))

    diff_mask = np.diff(x)
    change_indices = np.where(diff_mask != 0)[0]

    run_starts = np.concatenate(([0], change_indices + 1))

    vals = x[run_starts]

    run_ends = np.concatenate((change_indices + 1, [len(x)]))
    counts = run_ends - run_starts

    return (vals, counts)

x_list = [2, 2, 2, 3, 3, 3, 5, 5]
x_np = np.array(x_list)

print(f"x = {x_list}")
print("List:", task4_list(x_list))
print("Numpy:", task4_numpy(x_np))
```
